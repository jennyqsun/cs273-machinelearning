## Question 2

```python
# run k-means when k=2, k=5, k=20 with different initializations 
ks = [2,5,20]
init = ['random','random','random','farthest','k++']
sumd =  np.zeros((3,5))
z= np.zeros((3,X.shape[0],5))
randnum = [5,10,100]
mu = [0]*3

for i, k in enumerate(ks):
    np.random.seed(randnum[i])
    mu[i] = np.zeros((k,X.shape[1],5))
    for j, ini in enumerate(init):
        z[i,:,j], mu[i][:,:,j], sumd[i][j] = ml.cluster.kmeans(X,K=k, init = ini)
        print('K: %d'%k, ini, 'score: %f'%sumd[i][j])
    ind = (np.where(sumd[i]==min(sumd[i])))[0]
    if len(ind) == 1:
        print('number %d generates the best score'% ind)
        plt.figure()
        ml.plotClassify2D(None, X, np.squeeze(z[i][:,ind]))
        plt.scatter(mu[i][:,:,ind][:,0], mu[i][:,:,ind][:,1], s=100, marker = 'x', facecolor ='red')
        
    else:
        print('number %s generates same best scores'% ind)
        plt.figure()
        ind = ind[0]
        ml.plotClassify2D(None, X, np.squeeze(z[i][:,ind]))
        plt.scatter(mu[i][:,:,ind][:,0], mu[i][:,:,ind][:,1], s=100, marker = 'x', facecolor ='red')
        
#         print('best score')
        
```

When k =2, two random initiliazations generate same score. When k =3, k++ is the best. When k =20, one of the randopm initialization works the best.

Differences:

k-means allow data points to move between clusters as the algorithm improves its central values, whereas agglomerative clustering does not.

## Question 1

```python
# subtract the mean of the face images, to make the data zero-mean
mu = np.mean(X, axis=0)

# X0  = X - np.tile(meanX, (4916,1))
X0 = X - mu
plt.figure()
fig, ax = plt.subplots(1,5,figsize = (30,15))

faces = np.arange(120,600,100)
for i, f in enumerate(faces):
    img = np.reshape(X0[f,:],(24,24))
    ax[i].imshow(img.T, cmap="gray")
```

## Question 2

```python
# use scipy.linalg.svd to take the SVD of the data
from scipy import linalg

U,S,Vh = linalg.svd(X0, full_matrices = False)
W = U.dot(np.diag(S))
print('shape of W:', W.shape)
print('shape of Vh:', Vh.shape)
```

## Question 3

```python
K = np.arange(1,13,1)
mse = []
for i, k in enumerate(K):
    X0hat = W[:,:k].dot(Vh[:k,:]) #approximation to x0 by using first k eigenvectors
    mse.append(np.mean((X0 - X0hat)**2))

plt.plot(K, mse)
```

## Question 4

```python
# Display the first three PCs of the data.
fix, ax = plt.subplots(3,2,figsize = (10,15))
pc = 3
for i in range(pc):
    alpha = 2*np.median(np.abs(W[:,i]))
    pc1 = np.reshape(mu+alpha*Vh[i,:],(24,24))
    pc2 = np.reshape(mu-alpha*Vh[i,:],(24,24))
    ax[i][0].imshow(pc1.T, cmap="gray")
    ax[i][0].set_title('PC %d'%i)
    ax[i][1].imshow(pc2.T, cmap="gray")
    ax[i][1].set_title('PC %d'%i)
```

## Question 5

```python
# choose any two faces and reconstruct them using the first K principal directions

# using face 1 and face 4

Ks = [5,10,50,100]
Face = [0,3]
fig, ax = plt.subplots(2,4, figsize=(15,10))
for i, f in enumerate(Face):
    for j, k in enumerate(Ks):
        Xre = W[f,:k].dot(Vh[:k,:])
        img = np.reshape(Xre, (24,24))
        ax[i][j].imshow(img.T, cmap="gray")
        ax[i][j].set_title('K = %d'%k)
```
